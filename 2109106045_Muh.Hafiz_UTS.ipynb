## a. Buat Tabel Distribusi Frekuensi dan Histogram Variabe reading_score dan writing_score dengan jumlah dan lebar interval sebesar 10

### Skor Membaca

```python
interval = pd.interval_range(0,100, freq=10)
frekuensi = pd.cut(df['reading_score'], interval)
frekuensi.value_counts(sort=False).rename_axis("Skor Membaca").reset_index(name="Frekuensi")
```

#### Skor Menulis

```python
interval = pd.interval_range(0,100, freq=10)
frekuensi = pd.cut(df['writing_score'], interval)
frekuensi.value_counts(sort=False).rename_axis("Skor Menulis").reset_index(name="Frekuensi")
```

## c. Tentukan nilai Variasi: Jangkauan Range, Kisaran Interkuartil, Varians, Standar Deviasi data reading_score dan writing_score

```python
a = df['reading_score'].max()
b = df['reading_score'].min()
j_range= a-b
interkuartil = df['reading_score'].quantile()
varians = df['reading_score'].var()
std_deviasi = df['reading_score'].std()
print("Nilai Variasi")
print("Jangkauan Range: ",j_range)
print("Kisaran Interkuartil: ",interkuartil)
print("Varians: ",varians)
print("Standar Daviasi: ",std_deviasi)
```
